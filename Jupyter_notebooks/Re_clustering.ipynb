```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('../Data_bases/data_for_clustering.csv')
```

```python
data.head()
```

Selecting the numericals for scaling:

```python
data_num = data.select_dtypes([np.number])
```

```python
data_num.head()
```

Using standar scaler to scale the numerical columns

```python
transformer = StandardScaler().fit(data_num)
data_num_scaled = transformer.transform(data_num)

data_num_scaled = pd.DataFrame(data_num_scaled, columns=data_num.columns)
```

```python
data_num_scaled.head()
```

Selecting features for the first cluster:

```python
data_perf = data_num_scaled[['processor','ram_(mb)','internal_storage_(gb)']]
```

Using kmeans and ploting interections in order to check for the best number of clusters for each of the 4 clusters:

### PERFORMANCE

```python
from sklearn.metrics import silhouette_score
K = range(2, 10)

inertia = []
silhouette = []

for k in K:
    print("Training a K-Means model with {} clusters! ".format(k))
    print()
    kmeans = KMeans(n_clusters=k,
                    random_state=12,
                    verbose=1)
    kmeans.fit(data_perf)
    
    inertia.append(kmeans.inertia_)
    silhouette.append(silhouette_score(data_perf, kmeans.predict(data_perf)))



import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(1,2,figsize=(16,8))
ax[0].plot(K, inertia, 'bx-')
ax[0].set_xlabel('k')
ax[0].set_ylabel('inertia')
ax[0].set_xticks(np.arange(min(K), max(K)+1, 1.0))
ax[0].set_title('Elbow Method showing the optimal k')
ax[1].plot(K, silhouette, 'bx-')
ax[1].set_xlabel('k')
ax[1].set_ylabel('silhouette score')
ax[1].set_xticks(np.arange(min(K), max(K)+1, 1.0))
ax[1].set_title('Silhouette Method showing the optimal k')
```

```python
model = KMeans(n_clusters=5)
perfomance_c = model.fit_predict(data_perf)
perfomance_c
data['performance_c'] = perfomance_c
```
