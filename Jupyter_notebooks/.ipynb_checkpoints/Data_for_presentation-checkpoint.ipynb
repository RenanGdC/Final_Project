```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../Data_bases/ndtv_data_final.csv')
```

```python
data = data.drop('Unnamed: 0', axis=1)
```

```python
cols = [] 
for col in data.columns: 
    cols.append(col.lower().replace(' ', '_')) 
data.columns = cols

data.head()
```

```python
data['price'] = data['price']*0.012
data['price'] = data['price'].round(2)
def format_value(val):
    return '{:.2f}'.format(val)

data['price'] = data['price'].map(format_value)

data['price'] = data['price'].astype(float)
```

```python
data = data.dropna(axis=1)
```

```python
bin_prices=[5.93, 200, 400, 600, 800, 1000, 2099.88]
bin_ranges=['<200€', '200-400€', '400-600€', '600-800€', '800-1000€', '1000+€']
data['price_ranges']=pd.cut(data['price'], bins = bin_prices,
                                               labels = bin_ranges)
```

```python
data['screen_size_(inches)'].describe()
```

```python
bin_size=[2.3, 3.0,4.0,5.0,6.0,7.3]
bin_size_range=['2.3-3.0(inches)', '3.0-4.0(inches)', '4.0-5.0(inches)', '5.0-6.0(inches)', '6+(inches)']
data['screen_size_range']=pd.cut(data['screen_size_(inches)'], bins = bin_size,
                                               labels = bin_size_range)
```

```python
data['battery_capacity_(mah)'].describe()
```

```python
bin_bat=[1010.0, 2300.0,3000.0,3500.0, 4500.0,6000.0]
bin_bat_range=['1010-2300(mah)', '2300-3000(mah)', '3000-3500(mah)', '3500-4500(mah)', '4500-6000(mah)']
data['battery_capacity_range']=pd.cut(data['battery_capacity_(mah)'], bins = bin_bat,
                                               labels = bin_bat_range)
```

```python
data['rear_camera'].describe()
```

```python
data['rear_camera'].unique()
```

```python
bin_cam=[0,24,32,48,64,108]
bin_cam_range=['0-24(mp)', '24-32(mp)', '32-48(mp)', '48-64(mp)', '64+(mp)']
data['rear_camera_range']=pd.cut(data['rear_camera'], bins = bin_cam,
                                               labels = bin_cam_range)
```

```python
data = data.dropna()
```

```python
data.head()
```

```python
data.to_csv('../Data_bases/data_for_presentation1.csv', index=False)
```

```python
data['brand'].unique()
```

```python
def clean_brand(x):
    keep_list = ['Samsung','Apple','LG','Motorola','Huawei','Google',
                 'Sony','Microsoft','Nokia','10.or', 'Alcatel','OnePlus']
    return x if x in keep_list else "other"

data["brand_p"] = data["brand"].apply(clean_brand)
```
