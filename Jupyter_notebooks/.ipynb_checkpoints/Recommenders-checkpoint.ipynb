```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Data_bases/data_clstered.csv')
```

```python
data.head()
```

```python
data['price'].describe()
```

```python
bin_prices=[5.93, 200, 400, 600, 800, 1000, 2099.88]
bin_ranges=['<200', '200-400', '400-600', '600-800', '800-1000', '1000+']
data['price_ranges']=pd.cut(data['price'], bins = bin_prices,
                                               labels = bin_ranges)
```

```python
bin_prices=[5.93, 200, 400, 600, 800, 1000, 2099.88]
bin_ranges=[1, 2, 3, 4, 5, 6]
data['price_range']=pd.cut(data['price'], bins = bin_prices,
                                               labels = bin_ranges)
```

```python
data.head()
```

```python
def recommender(performance, camera_q, battery_l, size, price_r):
    result = data

    # result[(result["performance_c"] == performance) & () & () & () & ()].sample()
    if performance == 1:
        x = result[result["performance_c"] > 5]
        result = pd.concat([result, x])
    if camera_q == 1:
        x = result[result["cam_c"] > 4]
        result = pd.concat([result, x])
    if battery_l == 1:
        x = result[result["battery_c"] > 5]
        result = pd.concat([result, x])
    if size == 1:
        x = result[result['size_c'] > 5]
        result = pd.concat([result, x])
    if price_r == 1:
        x = result[result['price_range'] < 4]
        result = pd.concat([result, x])

    return result
```

```python
in_perf = int(input("Does performance matters? [0|1]: "))
in_cam  = int(input("Does cam quality matter? [0|1]: "))
in_size = int(input("Does size matter? [0|1]: "))
in_battery = int(input("Does battery matter? [0|1]: "))
in_price = int(input("Does the price matter? [0|1]: "))
res = recommender(in_perf, in_cam, in_size, in_battery, in_price)
for i in range(5):
    model = res.iloc[i]['model']
    brand = res.iloc[i]['brand']
    print(f"Recommendation {i+1}: {brand} {model}")

```

```python
def recommender_2(performance, camera_q, battery_l, size): 
    result = data[(data["performance_c"] >= performance) & (data["cam_c"] >= camera_q) & 
                  (data["battery_c"] >= battery_l) & (data["size_c"] >= size)] 
    if len(result) == 0: 
        return "No phones found that match the input criteria" 
    else: 
        return result.sample(5) if len(result) >= 5 else result.sample(len(result))

in_perf = int(input("How important is performance for you? [1-6]: ")) 
in_cam = int(input("How important is camera quality for you? [1-7]: ")) 
in_size = int(input("How important is size for you? [1-8]: ")) 
in_battery = int(input("How important is the battery duration for you? [1-8]: ")) 
#in_price = input("Choose a price range [<200, 200-400, 400-600, 600-800, 800-1000, 1000+]: ")

res = recommender_2(in_perf, in_cam, in_size, in_battery).sort_values(by="price")

for i in range(len(res)):
    model = res.iloc[i]['model']
    brand = res.iloc[i]['brand']
    price = res.iloc[i]['price']
    print(f"Recommendation {i+1}: {brand} {model} {price}€")
```
