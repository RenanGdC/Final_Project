```python
import pandas as pd
import numpy as np
import streamlit as st
```

```python
data = pd.read_csv('../Data_bases/data_re_clustered.csv')
```

Checking the data:

```python
data.head()
```

```python
def build_amazon_url(data):
    df = data.copy()
    df['url'] = "https://www.amazon.com/s?k=" + df['brand'] + " " + df['model']
    df['url'] = df['url'].str.replace(' ', '+')
    return df
```

```python
data = build_amazon_url(data)
```

```python
data.head()
```

Defining a function to find the values of the clusters by filtering the values on the data base:

```python
def recommender_2(performance, camera_q, battery_l, size): 
    result = data[(data["performance_c"] >= performance) & (data["cam_c"] >= camera_q) & 
                  (data["battery_t"] <= battery_l) & (data["size_c"] >= size)] 
    if len(result) == 0:
        return result
    else: 
        return result.sample(5) if len(result) >= 5 else result.sample(len(result))
```

Importing the inputs from the user:

```python
in_perf = int(input("How important is performance for you? [1-5]: ")) 
in_cam = int(input("How important is camera quality for you? [1-3]: ")) 
in_size = int(input("How important is size for you? [1-3]: ")) 
in_battery = int(input("How important is the battery duration for you? [1-5]: ")) 

res = recommender_2(in_perf, in_cam, in_battery, in_size)
if not len(res):
    print("No phones found")
else:
    res = res.sort_values(by="price")

    for i in range(len(res)):
        model = res.iloc[i]['model']
        brand = res.iloc[i]['brand']
        price = res.iloc[i]['price']
        url = res.iloc[i]['url']
        print(f"Recommendation {i+1}: {brand} {model} {price}€ {url}")
```
